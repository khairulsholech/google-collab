```python
# import library yang dibutuhkan
import vega_datasets
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np 
from sklearn import svm
```

```python
iris = vega_datasets.data.iris()
```

```python
#karena tidak perlu memakai virginica maka virginica kita drop/hapus
iris = iris.drop(iris[iris['species'] == 'virginica'].index)
```

```python
# untuk melihat data yang kita pakai
iris.head()
```

```python
# untuk mengecheck iris apa saja yang ada didalam data
set(iris['species'])
```

```python
# untuk visualisasi dari data 
sn.pairplot(iris,hue='species')
```

```python
# untuk clasifier
clf = svm.SVC(kernel='linear')
# untuk targetnya
X = iris[['sepalWidth','petalWidth']]
y = iris['species']
clf.fit(X,y)
```

```python
# untuk setting warna pada species 
Colors = {'setosa': 'b', 'versicolor': 'r'}
# untuk visualisasi dari lebar sepal dan petal 
plt.scatter(iris['sepalWidth'],iris['petalWidth'],c=[colors[r] for r in iris ['species']])
```

```python
# membuat prediksi
clf.predict([[3.0,0.5]])
```

```python
Colors = {'setosa': 'b', 'versicolor': 'r'}
plt.scatter(iris['sepalWidth'],iris['petalWidth'],c=[Colors[r] for r in iris ['species']])

# trick untuk membuat garis pada visualisasinya
# untuk mengambil axis yang ada
ax = plt.gca() 
# untuk mengambil limit x dari grafik diatas
xlim = ax.get_xlim()   
# untuk mengambil limit y dari grafik diatas
ylim = ax.get_ylim()   

# untuk linier space terdiri dari sekumpulan elemen yang disebut titik, dan sekumpulan elemen yang disebut garis.
# membuat linier space dari limit x 0 sampai 1 diambil 30 titik.
xx = np.linspace(xlim[0],xlim[1],30)    
# untuk x nya di bagi 30 , y dibagi 30, angka tsb akan menjadi array,
yy = np.linspace(ylim[0],ylim[1],30)
 

# untuk arraynya di buat grid spaya bisa membuat counterplot (3D yang di visualisasinya jadi 2D)
XX,YY = np.meshgrid(xx,yy)

# untuk xx ravel nya membuka /reflat/tidak ada dimensinya, yy juga, supaya jadi nilai yang rata
xy = np.vstack([XX.ravel(),YY.ravel()]).T 
# pada clasifier xy di reshape menjadi XX   
Z = clf.decision_function(xy).reshape(XX.shape) 

# untuk visualisasi
ax.contour(XX,YY,Z, colors='k',levels = [-1,0,1], alpha=0.5, linestyles=['--','-','--'])  
```
