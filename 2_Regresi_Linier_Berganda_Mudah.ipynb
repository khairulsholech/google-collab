```python
# melakukan import dataset dan library lain untuk digunakan
import vega_datasets
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn import linear_model
import sklearn.metrics as sm
```

```python
# mengambil dataset dengan data cars()
cr = vega_datasets.data.cars()
# menampilkan pair plot (varian plots)
sn.pairplot(cr, size = 2)
```

```python
# check null
set(cr['Displacement'].isna())
```

```python
set(cr['Horsepower'].isna())
```

```python
set(cr['Miles_per_Gallon'].isna())
```

```python
set(cr['Weight_in_lbs'].isna())
```

```python
# mengisi nilai null
cr['Miles_per_Gallon'] = cr['Miles_per_Gallon'].fillna(cr['Miles_per_Gallon'].mean())
cr['Horsepower'] = cr['Horsepower'].fillna(cr['Horsepower'].mean())
```

```python
# membuat model linear regression
lin_reg = linear_model.LinearRegression()
```

```python
# insert kedalam scikit learn
lin_reg.fit(cr[['Displacement', 'Horsepower', 'Weight_in_lbs']], cr['Miles_per_Gallon'])
```

```python
# describe data
cr['Displacement'].describe()
```

```python
# describe data
cr['Horsepower'].describe()
```

```python
# membuat sebuah prediksi
lin_reg.predict([[100, 200, 4000]])
```
