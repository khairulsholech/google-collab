Jawab :
data mining.jpg

**Data mining** adalah metode dalam ilmu komputer yang biasa digunakan dalam proses pencarian knowledge. Tahapan di dalamnya berguna untuk mencari pola-pola tertentu dari data yang ada pada database. Biasanya, metode ini banyak ditemukan pada bidang machine learning dan statistika.

Pada awalnya, metode penambangan data dikembangkan karena kompleksitas kerja komputer yang semakin meningkat. Namun, disinilah keuntungan adanya data mining adalah proses pengumpulan dan seleksi data yang lebih praktis.

Serangkaian proses tahapan data mining tersebut memiliki tahap sebagai berikut (Tan, 2004):


1.   Pembersihan data (untuk membuang data yang tidak konsisten dan noise)
2.   Integrasi data (penggabungan data dari beberapa sumber)
3.   Transformasi data (data diubah menjadi bentuk yang sesuai untuk di-mining)
4.   Aplikasi teknik Data Mining, proses ekstraksi pola dari data yang ada
5.   Evaluasi pola yang ditemukan (proses interprestasi pola menjadi pengetahuan yang dapat digunakan untuk mendukung pengambilan keputusan)
6.   Presentasi pengetahuan (dengan teknik visualisasi)

(Gunakan Input dan Output Keyboard) (Bahasa Pemrograman Python)

Buatlah program dengan ketentuan sebagai berikut: Menentukan Harga setelah di diskon 25% dari harga barang senilai Rp.10.000.000

```python
#Membuat inputan nilai
harga_barang = int(input("masukan harga barang "))
# ketentuan diskon
potongan = int(input("masukan nilai diskon "))/100
# melihat nilai diskon
nilai_potongan = potongan * harga_barang
# melihat harga setelah dapat diskon
harga_akhir = harga_barang - nilai_potongan
# tampilkan output
print("potongan diskon : {} \nharga setelah diskon : {}".format(nilai_potongan,harga_akhir))
```

(Gunakan Input dan Output Keyboard) (Bahasa Pemrograman Python)

Tentukan besarnya potongan dari pembelian barang yang diberikan seorang pembeli, dengan kriteria :

a. Tidak ada potongan jika total pembelian kurang dari Rp. 100.000,-

b. Jika total pembelian lebih dari atau sama dengan Rp. 100.000,- 

potongan yang diterima sebesar 20% dari total pembelian.


```python
#Membuat inputan nilai
harga_barang = int(input("masukan harga barang : "))
# ketentuan diskon
if harga_barang >= 100000:
  # potongan yang diterima sebesar 20% dari total pembelian.
  potongan = 0.2*harga_barang
  # melihat harga setelah dapat diskon
  harga_akhir = harga_barang - potongan
  # tampilkan output
  print("potongan diskon : {} \nharga yang harus dibayar setelah diskon : {}".format(potongan,harga_akhir))   
else:
  # tampilkan output
  print("harga yang harus dibayar adalah : {}".format(harga_barang)) 
```
