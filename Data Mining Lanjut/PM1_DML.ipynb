```python
n = int(input("Ketik sebuah angka, dan faktorialnya akan diprinted: "))

if n < 0:
  raise ValueError("Anda harus memasukan bilangan bulat bukan negatif")

faktorial = 1

for i in range(2, n+1):
  faktorial *=i

print(faktorial)
```
