```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def predict_diabetes(usia, bb, tb):
    # Normalisasi input
    usia = (usia - 40) / 20
    bb = (bb - 60) / 20
    tb = (tb - 160) / 20
    
    # Menghitung nilai z berdasarkan faktor usia, berat badan, dan tinggi badan
    z = 0.5 * usia - 0.3 * bb + 0.1 * tb
    # Menghitung probabilitas dengan menggunakan fungsi sigmoid
    prob = sigmoid(z)
    
    # Jika probabilitas lebih besar dari threshold, maka prediksi positif (menderita diabetes)
    threshold = 0.4
    if prob > threshold:
        return "Positif"
    else:
        return "Negatif"
    
# Contoh penggunaan
usia = float(input("Masukkan usia: "))
bb = float(input("Masukkan berat badan (kg): "))
tb = float(input("Masukkan tinggi badan (cm): "))

print("Prediksi diabetes: ", predict_diabetes(usia, bb, tb))
# Plot grafik fungsi sigmoid
z = np.linspace(-10, 10, 100)
plt.plot(z, sigmoid(z))
plt.title("Grafik Sigmoid Function")
plt.xlabel("Nilai z")
plt.ylabel("Probabilitas - sigmoid(z)")
plt.show()
```
