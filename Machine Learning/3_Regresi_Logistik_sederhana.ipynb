```python
# Library-library yang diperlukan
import vega_datasets
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
# Membuat object iris dari data iris vega datasets
iris = vega_datasets.data.iris()
```

```python
# Menampilkan pair plot (varian plots)
sn.pairplot(iris, hue = 'species')
```

```python
# Menentukan category dengan perbedaan warna
categ = {'setosa': 0, 'versicolor': 1, 'virginica': 2}
colors = {'setosa': 'blue', 'versicolor': 'red', 'virginica': 'green'}

# Menentukan X & Y
plt.scatter(iris['petalWidth'],iris['petalWidth'],c=[colors[r] for r in iris['species']])
```

```python
# Drop versicolor
iris = iris.drop(iris[iris['species'] == 'versicolor'].index)
```

```python
# Drop versicolor
iris = iris.drop(iris[iris['species'] == 'versicolor'].index)
plt.scatter(iris['petalWidth'], [categ[r] for r in iris['species']], c=[colors[r] for r in iris['species']])
```

```python
# Menentukan regresi logistik
clf = LogisticRegression()
clf.fit(iris[['petalWidth']], iris['species'])
```

```python
# Mencari plot
clf.predict([[0.7]])
```

```python
# Dibawah satu masuk setosa, diatas satu masuk versicolor
clf.predict([[1.2]])
```
